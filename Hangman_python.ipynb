```python
from random import choice
def word_generator():
    word_list=["calicut","kollam","munnar","ernankulam","idukki","pottemel"]
    return choice(word_list)
word=word_generator()
guessed=" "
turns=10
while(True):
    print("you are left with {} turns ".format(turns))
    inp=input("make a guess:")
    turns=turns-1
    if inp in word:
        guessed=guessed+inp
    ungessed_char=0
    for i in word:
        if i in guessed:
            print(i,end="")
        else:
            ungessed_char+=1
            print("*",end="")
    if ungessed_char==0:
        print("\n YOU WON..")
        break
    if turns==0:
        print("\ntoin toin toin ... you ran out the attempts")
        break
```
